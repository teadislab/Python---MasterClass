```python
import psycopg2
```

```python
conn = psycopg2.connect(
    host="localhost",
    database="helloWorld",
    user="postgres",
    password="root")
```

```python
cur = conn.cursor()

# execute a statement
print('PostgreSQL database version:')
cur.execute('SELECT version()')

# display the PostgreSQL database server version
db_version = cur.fetchone()
print(db_version)

# close the communication with the PostgreSQL
cur.close()
```
